```python
# Import Libraries
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../Datasets/Titanic.csv')
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Box plot
sns.boxplot(data=data, x="Sex", y="Age", hue="Survived")
plt.title("Age distribution by gender and survival status")
plt.show()
```
