```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../Datasets/Titanic.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
m1 = data['Age'].mean()
m2 = data['Cabin'].mode()[0]
m3 = data['Embarked'].mode()[0]
```

```python
data['Age']=data['Age'].fillna(m1)
data['Cabin']=data['Cabin'].fillna(m2)
data['Embarked']=data['Embarked'].fillna(m3)
```

```python
data.isnull().sum()
```

```python
# Line plot
sns.lineplot(data=data,x='Pclass',y='Fare',marker='o')
```

```python
# scatter plot
sns.scatterplot(data=data,x='Age',y='Fare',hue="Sex")
```

```python
# BarPlot
sns.barplot(data=data,x='Pclass',y='Fare')
```

```python
# Box Plot
sns.boxplot(data=data,x='Pclass',y='Fare')
```

```python
# Histogram
sns.histplot(data=data,x=data['Fare'],bins=20,kde=True)
```

```python
# KDE PLOT
sns.kdeplot(data=data,x=data['Fare'],fill=True)
```

```python
# Box Plot
plt.figure(figsize=(8, 4))
sns.boxplot(data=data,x='Sex',y='Age',hue="Survived")
```
