### Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/Iris.csv')
data.head()
```

```python
sample_data = data.head(20)
```

### Line Plot

```python
sns.lineplot(x="sepal_length",y="petal_length",data=sample_data,hue="species")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

### Bar Plot


```python
sns.barplot(x="sepal_length", y="species", data=data)
plt.title("Sepal Length vs Species")
plt.show()
```

### Box plot

```python
sns.boxplot(x="species", y="sepal_length", data=data,hue="species",palette="Accent")
plt.title("Species wise Sepal Length")
plt.show()
```

### Scatter Plot

```python
sns.scatterplot(data=data, x="petal_length", y="petal_width", hue="species", style="species")
plt.title("Petal Length vs Petal Width")
plt.show()
```

### Count Plot

```python
sns.countplot(x='species', data=data, palette='Set2')
plt.title('Count of Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

### Histogram

```python
# Histogram of sepal_length
sns.histplot(data=data, x='sepal_length', bins=10, kde=True, color='green')
plt.title('Distribution of Sepal Length (Histogram)')
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.show()
```
